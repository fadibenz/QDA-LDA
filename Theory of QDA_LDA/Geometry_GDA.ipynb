# Notes on the Multivariate Normal Distribution



- The **normal distribution** is awesome, it's so simple yet so powerful and finds use cases everywhere in science but the premise is very simple.

- The **probability density function (PDF)** of the normal distribution is given by the formula:

  $$
  f(x) = \frac{1}{\sqrt{2\pi\sigma^2}} \exp\left(-\frac{(x-\mu)^2}{2\sigma^2}\right)
  $$

  where:
  - **$\mu$** represents the mean of the distribution, which is the central value around which the data points are expected to cluster.
  - **$\sigma^2$** is the variance, a measure of how spread out the data points are around the mean.

- The **probability density** is highest at the mean, reflecting the fact that values close to the mean are the most likely to occur. This is intuitive because if most values were far from the mean, the mean wouldn't truly represent the "center" of the data. As we move away from the mean, the probability density decreases, meaning values farther from the mean are less likely, which aligns with common sense.
One way I like to phrase it is the probability of being big (with respect to the mean) can not be too big.

- The rate at which the **probability density decreases** as we move away from the mean is determined by the variance.

- All of this is captured in the **argument of the exponential function** in the PDF, $\frac{(x-\mu)^2}{2\sigma^2}$. We calculate the squared distance from the mean to account for negative distances (We can use the absolute value, but integrating it...) we then penalize deviation from the mean by dividing with **$\sigma^2$**.

- If the data is more spread out, **$\sigma^2$** would be big, far distances from the mean will get smaller when we divide by **$\sigma^2$** and therefore will have higher probabilities compared to if **$\sigma^2$** was small and we had a big distance it would still be comparably big and the probability small. Think about it. 

- The rest of the formula is easy.
  
-  The **denominator in the PDF** ($\sqrt{2\pi\sigma^2}$) ensures the area under the curve of the PDF equals one, which is a requirement for all probability density functions.

- The function $\exp(-x^2)$ creates the characteristic bell shape of the normal distribution, which is symmetric around the mean. Many functions give the same shape and can therefore be used instead but the exponential integrates beautifully compared to other functions. Watch 3Blue1Brown video on the Normal Distribution.

- The distance calculated by $\frac{(x-\mu)^2}{2\sigma^2}$ is called the **Mahalanobis distance** between the value $x$ and the mean $\mu$, scaled by the variance $\sigma^2$. The Mahalanobis distance is a crucial metric because, unlike the Euclidean distance, it accounts for the spread of the distribution.

- Moving on to the **multivariate normal distribution**, which generalizes the univariate normal distribution to multiple dimensions. But before diving into the full generality, it's helpful to understand a special case known as the **isotropic normal distribution**. The PDF for the isotropic normal distribution is:

  $$
  f(\mathbf{x}) = \frac{1}{(2\pi\sigma^2)^{n/2}} \exp\left(-\frac{(\mathbf{x}-\mathbf{\mu})^T(\mathbf{x}-\mathbf{\mu})}{2\sigma^2}\right)
  $$

  where:
  - **$\mathbf{\mu}$** is the mean vector, representing the central point in the distribution across multiple dimensions.
  - **$\mathbf{x}$** is a vector of random variables, representing a point in multidimensional space.
  - **$\sigma$** is a scalar representing the standard deviation, which is the same across all directions (hence "isotropic," meaning uniform in all directions).
  - **$n$** is the number of dimensions.

- When dealing with multivariable functions, it's common to represent them using **isocontours** (Watch: Khan Academy multivariable calculus series). For the isotropic normal distribution, the isocontours are particularly simple and intuitive: they are spheres (or circles in two dimensions) centered at the mean vector $\mathbf{\mu}$. The radius of each sphere is determined by the variance $\sigma^2$. This spherical shape is a direct consequence of the fact that the variance is the same in all directions.

- The **isocontour equation** for the isotropic normal distribution is given by:

  $$
  \frac{1}{(2\pi\sigma^2)^{n/2}} \exp\left(-\frac{(\mathbf{x}-\mathbf{\mu})^T(\mathbf{x}-\mathbf{\mu})}{2\sigma^2}\right) = c
  $$

  Simplifying this equation reveals that the isocontours are indeed spheres. The center of each sphere is at the mean vector $\mathbf{\mu}$, and the radius of the sphere is related to the variance. If the variance were different in each direction, the shape of the isocontours would change; instead of spheres, they would be ellipsoids, reflecting the different spreads along each axis.

- Finally, let's consider the **general multivariate normal distribution**, which extends the isotropic case to allow for different variances in different directions and correlations between dimensions. The PDF for the multivariate normal distribution is:

  $$
  f(\mathbf{x}) = \frac{1}{\sqrt{(2\pi)^n|\boldsymbol{\Sigma}|}} \exp\left(-\frac{1}{2}(\mathbf{x}-\boldsymbol{\mu})^T \boldsymbol{\Sigma}^{-1} (\mathbf{x}-\boldsymbol{\mu})\right)
  $$

  where:
  - **$\mathbf{x}$** is the vector of random variables.
  - **$\boldsymbol{\mu}$** is the mean vector.
  - **$\boldsymbol{\Sigma}$** is the covariance matrix, which captures the variances of the individual variables (diagonal elements) and the covariances (off-diagonal elements) between pairs of variables.
  - **$n$** is the number of dimensions.

- The key difference between the **isotropic** and **multivariate normal distribution** is the use of the covariance matrix $\boldsymbol{\Sigma}$ instead of a single scalar variance $\sigma^2$. This matrix allows the multivariate normal distribution to model the dependencies and correlations between different dimensions. In the isotropic case, all dimensions are treated the same, but in the multivariate case, each dimension can have a dif.

- Since there are diffrent carainces in each axis we will have elipsoids. t variance, and there can be correlations between dimensions.

- The ellipsoids are sometimes also tilted; this is because our variables can be correlated. The tilt represents the slope of the correlation (Watch: statquest Covariance).

- Now for a deeper look at this geometry, let's do the same as before: put the expression of the Multivariate Normal equal to a value c (So we can see the isocontour for the isovalue c). After simplification, we will get:
  $(\mathbf{x}-\boldsymbol{\mu})^T \boldsymbol{\Sigma}^{-1} (\mathbf{x}-\boldsymbol{\mu})$ = c
This is a quadratic form; it represents an ellipsoid (Read: Gilbert Strang, Introduction to Linear Algebra, Fifth edition).

- So contours of a quadratic expression $(\mathbf{x}-\boldsymbol{\mu})^T \boldsymbol{M}(\mathbf{x}-\boldsymbol{\mu})$ are ellipsoids determined by eigenvectors/eigenvalues of $\boldsymbol{M}^{-1/2}$ (Read notes on visualizing quadratic form).

- Now for our precision matrix $\boldsymbol{\Sigma}^{-1}$ (which determines the shape of our multivariate normal, since the rest are just scalars), the ellipsoids are determined by eigenvectors/eigenvalues of $\boldsymbol{\Sigma}^{1/2}$, which is the standard deviation. (This is why when we whiten data we multiply by $\boldsymbol{\Sigma}^{-1/2}$, to reverse this effect, we map ellipsoids to spheres)

- The longer axis is determined by the maximum standard deviation, which corresponds to the largest eigenvalue of $\boldsymbol{\Sigma}$. Points in this direction are less penalized since data is more spread out.

## Isocountours of Normal Distribution

In the rest of this notebook I visualize some distributions with additional notes.

```python
import sys
if sys.version_info[0] < 3:
	raise Exception("Python 3 not detected.")
import numpy as np
import scipy.stats
import matplotlib.pyplot as plt


plt.figure(figsize=(8, 8))
x = np.linspace(-2, 4, 500)
y = np.linspace(-2, 4, 500)
X, Y = np.meshgrid(x, y)
posT = np.array([Y, X]).T
rv = scipy.stats.multivariate_normal([1, 1], [[1, 0], [0, 1]])
Z = rv.pdf(posT)
plt.contourf(X, Y, Z, 10)
plt.title('Isocontour of distribution')
plt.colorbar()
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

In this first iscontour we notice our normal distribution is spheres, this is becuase the variance of our two random variables is equal, so the eigenvalues of our covariance matrix (diagonal entries) are equal,  which means distance is penalized the same in both directions, no axis is longer than the other, we get spehres instead of ellipsoids.

```python
plt.figure(figsize=(8, 8))
x = np.linspace(-4, 4, 500)
y = np.linspace(-4, 4, 500)
X, Y = np.meshgrid(x, y)
posT = np.array([Y, X]).T
rv = scipy.stats.multivariate_normal([-1, 2], [[2, 0], [0, 1]])
Z = rv.pdf(posT)
plt.contourf(X, Y, Z)
plt.title('Isocontour of distribution')
plt.colorbar()
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

Here it's similliar to the previous case, off-diagonal entries are zero, but we have diffrent diagonal entries, diffrent eigenvalues, meaning diffrent variance for each variable. the distance is then penalized diffrently, which expalins the elipsoidal shape. The longer axis coressponds to the larger eignevalue which corespondes to the smallest eignevalue of the precision matrix which eplains why variance is less penalized in this direction.

We might also notice that our iscountour is not tilted, this is because our varibles are not corellated, correlation geometrically is represnted by the tilt of our elipsoids, which means a positive or negative slope (positive or negative correlation)

```python

plt.figure(figsize=(8, 8))
x = np.linspace(-4, 5, 500)
y = np.linspace(-4, 5, 500)
X, Y = np.meshgrid(x, y)
posT = np.array([Y, X]).T
rv1 = scipy.stats.multivariate_normal([0, 2], [[2, 1], [1, 1]])
rv2 = scipy.stats.multivariate_normal([2, 0], [[2, 1], [1, 1]])
Z1 = rv1.pdf(posT)
Z2 = rv2.pdf(posT)
plt.contourf(X, Y, Z1 - Z2)
plt.title('Isocontour of distribution')
plt.colorbar()
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

This is the diffrence between two normal distributions, we can see the bayes optimal decision boundary, which is linear in this case, LDA would do a good job estimating this boundary, Although we would never find the real normal distribution parameteres.

# Eigenvectors of the Gaussian Covariance Matrix

```python
import sys
if sys.version_info[0] < 3:
	raise Exception("Python 3 not detected.")
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm
import scipy
import numpy as np
import scipy.stats
import matplotlib.pyplot as plt

#Sample points
mu_X1 = 3
sigma_X1 = 3
mu_X2 = 4
sigma_X2 = 2
n_samples = 100
np.random.seed(9)
samples_X1 = np.random.normal(mu_X1, sigma_X1, n_samples)
samples_X2_normal = np.random.normal(mu_X2, sigma_X2, n_samples)
samples_X2 = 0.5 * samples_X1 + samples_X2_normal

sample = np.array([samples_X1, samples_X2]).T


#Mean
sample_mean = np.mean(sample, axis=0)

#Covariance
sample_cov = np.cov(sample, rowvar=False)


#Eigenthings
eigenvalues, eigenvectors = np.linalg.eig(sample_cov)
plt.figure(figsize=(8, 8))
plt.scatter(samples_X1, samples_X2)

plt.xlim(-15, 15)
plt.ylim(-15, 15)
plt.gca().set_aspect('equal', adjustable='box')

vec_X = [sample_mean[0], sample_mean[0]]
vec_Y = [sample_mean[1], sample_mean[1]]
vec_U = [eigenvectors[0][0] * eigenvalues[0], eigenvectors[0][1] * eigenvalues[1]]
vec_V = [eigenvectors[1][0] * eigenvalues[0], eigenvectors[1][1] * eigenvalues[1]]
plt.quiver(vec_X, vec_Y, vec_U, vec_V, angles="xy", scale_units="xy", scale=1)
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('Sample points and Eigenthings')


plt.show()


rotated_samples = (eigenvectors.T @ (sample - sample_mean).T).T

plt.figure(figsize=(8, 8))
plt.scatter(rotated_samples[:, 0], rotated_samples[:, 1])


plt.xlim(-15, 15)
plt.ylim(-15, 15)
plt.gca().set_aspect('equal', adjustable='box')
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('Sample points rotated')


plt.show()
```

Here we can see how the eigenvalues of our covariance matrix are coordinate-axis.

When we rotate our sample points, they are decorrelated, our normal distribution follows a normal distribution of 0-mean and a diagonal covariance matrix.

if the diagonal matrix is the identity matrix, our data is whitened.
